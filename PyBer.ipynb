## Merge DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how='left', on=["city", "city"])

# Display the DataFrame
pyber_data_df.head()
```

## Create seperate DataFrame for each city type 

```python
# Create the urban city DataFrame

urban_cities_df = pyber_data_df[pyber_data_df['type'] == 'Urban']
urban_cities_df.head()
```

```python
# Create the suburban city DataFrame

suburban_cities_df = pyber_data_df[pyber_data_df['type'] == 'Suburban']
suburban_cities_df.head()
```

```python
# Create the rural city DataFrame

rural_cities_df = pyber_data_df[pyber_data_df['type'] == 'Rural']
rural_cities_df.head()
```

## Create the number of rides for each type of city - X axis

```python
# Get the number of rides for urban cities

urban_ride_count = urban_cities_df.groupby(['city']).count()['ride_id']
urban_ride_count.head()
```

```python
# Get the number of rides for suburban cities

suburban_ride_count = suburban_cities_df.groupby(['city']).count()['ride_id']
suburban_ride_count.head()
```

```python
# Get the number of rides for rural cities

rural_ride_count = rural_cities_df.groupby(['city']).count()['ride_id']
rural_ride_count.head()
```

## Create the average fare for each type of city - Y axis

```python
# Get average fare for each city in the urban cities

urban_avg_fare = urban_cities_df.groupby('city').mean()['fare']
urban_avg_fare.head()
```

```python
# Get average fare for each city in the suburban cities

suburban_avg_fare = suburban_cities_df.groupby('city').mean()['fare']
suburban_avg_fare.head()
```

```python
# Get the average fare for each city in the rural cities

rural_avg_fare = rural_cities_df.groupby('city').mean()['fare']
rural_avg_fare.head()
```

## Create the average number of drivers for each type of city - size of the scatter plot

```python
# Get the average number of drivers for each urban city.

urban_driver_count = urban_cities_df.groupby('city').mean()['driver_count']
urban_driver_count.head()
```

```python
# Get the average number of drivers for each suburban city.

suburban_driver_count = suburban_cities_df.groupby('city').mean()['driver_count']
suburban_driver_count.head()
```

```python
# Get the average number of drivers for each rural city.

rural_driver_count = rural_cities_df.groupby('city').mean()['driver_count']
rural_driver_count.head()
```

## Create the Urban cities Bubble Chart

```python
# Create the scatter plots for urban cities

plt.scatter(urban_ride_count,
        urban_avg_fare,
        s=10 * urban_driver_count, c='coral',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Urban')
          
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total number of Rides (Per City)")
plt.grid(True)

# Add the legend
plt.legend()
```

## Create the Suburban Cities Bubble Chart

```python
# Build the scatter plot for the suburban cities

plt.scatter(suburban_ride_count,
        suburban_avg_fare,
        s=10 * suburban_driver_count, c='skyblue',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Suburban')

plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total number of Rides (Per City)")
plt.grid(True)

# Add the legend
plt.legend()
```

## Create the Rural Cities Bubble Chart

```python
# Build the scatter plot for the rural cities

plt.scatter(rural_ride_count,
        rural_avg_fare,
        s=10 * rural_driver_count, c='gold',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Rural')

plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total number of Rides (Per City)")
plt.grid(True)

# Add the legend
plt.legend()
```

## Combine all three scatter plots in one 

```python
# Add the scatter charts for each city type

plt.scatter(urban_ride_count,
        urban_avg_fare,
        s=10 * urban_driver_count, c='coral',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Urban')

plt.scatter(suburban_ride_count,
        suburban_avg_fare,
        s=10 * suburban_driver_count, c='skyblue',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Suburban')

plt.scatter(rural_ride_count,
        rural_avg_fare,
        s=10 * rural_driver_count, c='gold',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Rural')

# Show the plot
plt.show()
```

```python
# Add the scatter charts for each city type with annotations

plt.subplots(figsize=(10, 6))

plt.scatter(urban_ride_count,
        urban_avg_fare,
        s=10 * urban_driver_count, c='coral',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Urban')

plt.scatter(suburban_ride_count,
        suburban_avg_fare,
        s=10 * suburban_driver_count, c='skyblue',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Suburban')

plt.scatter(rural_ride_count,
        rural_avg_fare,
        s=10 * rural_driver_count, c='gold',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Rural')

# Incorporate the other graph properties
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.xlabel("Total number of Rides (Per City)", fontsize=12)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.grid(True)

# Add the legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Add the scatter charts for each city type with annotations and create a legend

plt.subplots(figsize=(10, 6))

plt.scatter(urban_ride_count,
        urban_avg_fare,
        s=10 * urban_driver_count, c='coral',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Urban')

plt.scatter(suburban_ride_count,
        suburban_avg_fare,
        s=10 * suburban_driver_count, c='skyblue',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Suburban')

plt.scatter(rural_ride_count,
        rural_avg_fare,
        s=10 * rural_driver_count, c='gold',
        edgecolor='k', linewidths=1,
        alpha=0.8, label='Rural')

# Incorporate the other graph properties
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.xlabel("Total number of Rides (Per City)", fontsize=12)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.grid(True)

# Create a legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
        scatterpoints=1, loc="best", title="City types")

lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Incorporate a text label about circle size
plt.text(42, 35, "Note: Circle size correlates with driver count per city.", fontsize="12")

# Save the figure
plt.savefig("analysis/Fig1.png", bbox_inches='tight')
```

## Get the Percentage of Fares for Each City Type

```python
# Get the sum of the fares for each city type.

sum_fares_by_type = pyber_data_df.groupby("type").sum()["fare"]
sum_fares_by_type
```

```python
# Get the sum of all the fares.

total_fares = pyber_data_df['fare'].sum()
total_fares
```

```python
# Calculate the percentage of fare for each city type.

type_percents = 100 * sum_fares_by_type / total_fares
type_percents
```

```python
# Alternate and concise code

type_percents = 100 * pyber_data_df.groupby('type').sum()['fare'] / pyber_data_df['fare'].sum()
type_percents
```

```python
type_percents.index
```

```python
# Build the percentage of fares by city type pie chart.

plt.pie(type_percents, 
    labels=type_percents.index,
    colors=["gold", "lightskyblue", "lightcoral"],
    explode=[0, 0, 0.1],
    autopct='%1.1f%%',
    shadow=True, startangle=150)
    
plt.title("% of Total Fares by City Type")
plt.axis('equal')
plt.show()
```

```python
# Import mpl to change the plot configurations using rcParams.

import matplotlib as mpl

#Build the pie chart
plt.subplots(figsize=(10, 6))
plt.pie(type_percents, 
    labels=type_percents.index,
    colors=["gold", "lightskyblue", "lightcoral"],
    explode=[0, 0, 0.1],
    autopct='%1.1f%%',
    shadow=True, startangle=150)

plt.title("% of Total Fares by City Type")

# Change the default font size from 10 to 14.
mpl.rcParams['font.size']= 14

plt.axis('equal')

# Save figure
plt.savefig("analysis/Fig5.png")

# Show figure
plt.show()
```
