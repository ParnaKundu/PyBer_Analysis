## Create a Line chart 

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot using MATLAB method

plt.plot(x_axis, y_axis)
```

```python
# Create the plot using Object_Oriented Interface 
# Create the plot with ax.plot()

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label="Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the plot and add a label for the legend.
ax = plt.axes()
ax.plot(x_axis, y_axis, color="green", marker="d", linewidth=2, label="Boston")

# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")

# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)

# Create a title.
ax.set_title("PyBer Fare by Month")

# Add a grid.
ax.grid()

# Add the legend.
ax.legend()
```

## Create a Bar chart

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create Vertical Bar chart using the MATLAB method.
plt.bar(x_axis, y_axis, color="green", label="Boston")

# Create labels for x and y axis.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot with the values in reverse axis
plt.barh(y_axis, x_axis)
```

```python
# Create the Horizontal Bar Chart with the y-axis data in reversed order
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Add annotation to a bar chart
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()

plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

plt.legend()
```

```python
# Create the bar chart using the Object-Oriented method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the bar chart using the Object-Oriented method with annotations
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="k", label="New York")
ax.legend()
```

```python
# Create the Horizontal Bar chart using the Object-Oriented method 
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the Horizontal Bar chart using the Object-Oriented method and switching the data on opposite axes
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Create the Horizontal Bar chart using the Object-Oriented method with annotations
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="c", label="Chicago")
ax.invert_yaxis()

ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

ax.legend()
```

## Create a Scatter plot

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create Scatter plot using MATLAB
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create a scatter plot
plt.scatter(y_axis, x_axis, c='r', label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.xlim(0, 45)
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Create a Bubble chart using MATLAB method
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create Bubble chart where the markers are 3 times it corresponding y-axis values

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Use list comprehension to determine size

plt.scatter(x_axis, y_axis, s = [ i*3 for i in y_axis])
```

```python
# Create scatter plot using Object-Oriented Method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble chart
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create a Bubble Chart - 20% transparent, size is 5 times data points, marker is skyblue, black edge color,
# linewidth of marker edge is 2 points

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,c='skyblue', s=[i*5 for i in y_axis], alpha=0.8, edgecolors='k', 
           linewidths=2, label='Boston')

ax.set_title('PyBer Fare by Month')


ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_xlim(0, 50)
ax.legend()

ax.invert_yaxis()
```

## Create a Pie Chart

```python
# Set the list of strings for each month as label.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the list of total fare in US dollars accumulated for each month as the values.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create pie chart using the MATLAB method
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Create pie chart using MATLAB method and add annotations
explode_values = [0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]

plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Create pie chart using MATLAB method and add annotations - increase the figure size
plt.subplots(figsize=(8, 8))
explode_values = [0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]

plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Create pie chart using MATLAB method and add annotations - change the color of each pie wedge

colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen', 'greenyellow', 'yellow', 'orange', 'gold', 
          'indianred', 'tomato', 'mistyrose']

explode_values = [0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]

plt.subplots(figsize=(8, 8))
plt.pie(y_axis, 
        explode=explode_values,
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')

plt.show()
```

```python
# Create pie chart using the Object-oriented Interface

fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Create pie chart using the Object-oriented Interface - add annotations


colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen', 'greenyellow', 'yellow', 'orange', 'gold', 
          'indianred', 'tomato', 'mistyrose']

explode_values = [0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%',
       shadow=True,
       startangle=90,
       counterclock=False)
plt.show()
```

## Add Error Bars to a line chart

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis

stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

## Add Error bar to Bar Chart

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

## Change the Major ticks

```python
import numpy as np
```

```python
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
